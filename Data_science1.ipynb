```python
Q1 = df_fish['Weight'].quantile(0.25)
Q3= df_fish['Weight'].quantile(0.75)
X = Q3 - Q1
low_bound = Q1 - 1.5 * X
up_bound = Q3 + 1.5 * X
plt.figure(figsize=(48, 18))
outliers = df_fish[(df_fish['Weight'] < low_bound) | (df_fish['Weight'] > up_bound)]
sns.boxplot(x='Species', y='Weight', data=df_fish)
plt.scatter(outliers.index, outliers['Weight'], color='blue', label='Weight_Outliers')
plt.xlabel('Species')
plt.ylabel('Weight')
plt.title('Box Plot for weight with Outliers')
plt.legend()
plt.show()
```
